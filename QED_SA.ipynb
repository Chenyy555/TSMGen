```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


df_3 = pd.read_csv('./save/pre/crossdocked/char/hgnn/2025_01_05_20/sample_300_30_True_1_1_1741635931/pocket_metrics.csv')
df_4 = pd.read_csv('./save/pre/crossdocked/char/all/2025_01_12_16/sample_300_30_True_1_1_1743065493/pocket_metrics.csv')
# save/pre/crossdocked/char/all/2025_01_12_16/sample_300_30_True_1_1_1743065493/pocket_metrics.csv

df_3_sa = df_3['sas0-1']
df_4_sa = df_4['sas0-1']


sns.kdeplot(df_3_sa, color='#1f77b4', shade=True, label='PHy2Mol')
sns.kdeplot(df_4_sa, color='#ff7f0e', shade=True, label='MB2Mol')
# plt.yticks([],fontsize=14)
plt.xlabel('SA',fontsize=12)
plt.legend(loc='upper right',fontsize=12)
plt.savefig('./figure/sa_1.png',dpi=500)
plt.show()
```
